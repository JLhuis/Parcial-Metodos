```python
import numpy as np
import matplotlib.pyplot as plt

# ── Datos experimentales ──────────────────────────────────────────────────────
f = np.array([100,120,145,170,200,235,270,310,355,405,460,520,585,655,730,
              810,895,985,1080,1180,1290,1410,1540,1680,1830,1990,2160,2340,2530,2730])
Z = np.array([152.3,149.1,146.8,144.9,142.0,139.5,137.9,136.1,134.8,133.6,
              132.7,131.9,131.4,131.1,130.9,131.0,131.3,131.9,132.7,133.8,
              135.2,136.9,138.9,141.1,143.5,146.1,149.0,152.2,155.6,159.2])

# ── Gráfico principal ─────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(f, Z, 'o-', color='steelblue', ms=5, lw=1.8, label='|Z|(f) medida')

# Estimar mínimo visualmente (valor mínimo en los datos)
idx_min = np.argmin(Z)
f_min, Z_min = f[idx_min], Z[idx_min]

# Marcar mínimo
ax.axvline(f_min, color='tomato', ls='--', lw=1.2, alpha=0.8)
ax.plot(f_min, Z_min, 'r*', ms=12, label=f'Mínimo ≈ {f_min} Hz  |Z| = {Z_min:.1f} Ω')

ax.set_xlabel('Frecuencia f (Hz)', fontsize=12)
ax.set_ylabel('|Z| (Ω)', fontsize=12)
ax.set_title('Impedancia bioeléctrica vs. Frecuencia (T = 37 °C)', fontsize=13)
ax.legend(fontsize=10)
ax.grid(True, alpha=0.35)
plt.tight_layout()
plt.savefig('parte_A_impedancia.png', dpi=150)
plt.show()

# ── Resultados en consola ─────────────────────────────────────────────────────
print(f"Mínimo local (datos crudos): f = {f_min} Hz,  |Z| = {Z_min:.4f} Ω")
print(f"Rango |Z|: [{Z.min():.4f}, {Z.max():.4f}] Ω")
print(f"Variación total: {Z.max()-Z.min():.4f} Ω")
```
